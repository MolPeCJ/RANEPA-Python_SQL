#### Задание 1.10

Перед вами набор чисел x. набор y задайте как список из нулей такой же длины. Это будут координаты центров кругов. Нарисуйте в этих точках круги с возрастающими радиусами, начните с радиуса 1 и  закончите радиусом 5.

```python
x = [2, 6, 12, 20, 30]
y = np.zeros(len(x))
```

```python
plt.figure(figsize = (10, 5))

plt.title('Задание 1.10')
plt.xlabel('Ось х')
plt.ylabel('Ось y')
plt.xlim(-40, 75)
plt.ylim(-10, 10) 

plt.axis('equal') # чтобы круги не были эллипсами

for i in range(len(x)):
    plt.gca().add_artist(plt.Circle((x[i], y[i]), radius = i + 1, fill = True)) # специальная функция для кругов

plt.show()
```
