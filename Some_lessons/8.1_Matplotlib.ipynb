### Matplotlib - библиотека для визуализации данных

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
x = [1, 2, 3, 10, 5]
y = [0, 7, 2, 10, 8]

plt.plot(x, y)
plt.show()
````

### Координатная сетка

```plt.grid(**kwargs)``` - нанести сетку с необязательными параметрами:
* linestyle - вид сетки ('--', ':', ...)
* linewidth - ширина линии
* color - цвет

````python
x = [1, 2, 3, 10, 5]
y = [0, 7, 2, 10, 8]

plt.plot(x, y)
plt.grid(color = 'g', ls = '--', linewidth = 0.5) # создание сетки и изменение ее параметров
````

### Масштабы и размеры

```plt.axes().set_aspect(1)``` - задает соотношение масштабов осей  
```plt.figure(figsize=(10, 5))``` - задает ширину и высоту полотна

````python
x = [1, 2, 3, 10, 5]
y = [0, 7, 2, 10, 8]

plt.axes().set_aspect(1) # задает соотношение масштабов осей
plt.figure(figsize = (10, 5)) # задает соотношение сторон - выбираем что-то одно

plt.plot(x, y)

plt.show()
````

### Диаграмма рассеяния

```plt.scatter(x, y, **kwargs)``` - рисует точки на графике
* s/sizes - размер/набор размеров
* c/color - цвет или набор цветов
* marker - вид точек: ('o', 's', '*', '+' и др)
* alpha - прозрачность
* cmap - используемая карта цветов ('viridis', 'plasma' и др)

````python
x = [1, 2, 3, 10, 5]
y = [0, 7, 2, 10, 10]

plt.scatter(x, y, sizes = [200, 10, 111, 40, 400], color = ['#af1500', '#ff1600', '#fa1700', '#ff1800', '#ff1900'], marker = '*')

plt.show()
````

### Уникальные цвета и размеры точек

````python
x = np.random.rand(100)
y = np.random.rand(100)

colors = np.random.randint(0, 5, 100)
sizes = 1000 * np.random.rand(100)

plt.scatter(x, y, c = colors, s = sizes, alpha = 0.5, cmap = 'plasma') # добавили прозрачность и выбрали цветовую палитру
plt.colorbar() # добавляет градиентную шкалу

plt.show()
````

### Графики функций

````python
def f(x):
    return x**2 + 2*x + 1

x = np.arange(-10, 10, 0.5) # random.rand, random.randint - не работают так
y = f(x)

plt.scatter(x, y)
plt.plot(x, y)

plt.show()
````

````python
def f1(x):
    return x**2

def f2(x):
    return x**2 + 5 * x

def f3(x):
    return x**2 + 10

x = np.arange(-10, 10, 0.5)
y1 = f1(x)
y2 = f2(x)
y3 = f3(x)

plt.plot(x, y1, 'b')
plt.plot(x, y2, 'r-.')
plt.plot(x, y3, 'g--')

plt.show()
````

### Легенда графика

Чтобы не искать, какая линия к какой функции относится, можно отобразить легенду.  
Для этого необходимо каждому графику задать метку ```label``` и вызвать функцию ```plt.legend()```

````python
plt.plot(x, y1, 'b', label = 'x**2')
plt.plot(x, y2, 'r-.', label = 'x**2 + 5 * x')
plt.plot(x, y3, 'g--', label = 'x**2 + 10')

plt.legend(fontsize = 14, shadow = True) # меняем размер легенды и добавляем эффект тени
plt.show()
````

### Оси графика

Для большей информативности можно обозначить названия осей:  
Подпись к оси Х: ```plt.xlabel('имя_оси', **kwargs)```  
Подпись к оси Y: ```plt.ylabel('имя_оси', **kwargs)```

````python
plt.plot(x, y1, 'b')
plt.plot(x, y2, 'r-.')
plt.plot(x, y3, 'g--')

plt.xlabel('Ось х')
plt.ylabel('Ось y')

plt.show()
````

Диапазон видимых значений осей устанавливается следующим образом:  
```plt.xlim(от, до)```  
```plt.ylim(от, до)```

````python
plt.plot(x, y1, 'b')
plt.plot(x, y2, 'r-.')
plt.plot(x, y3, 'g--')

plt.xlim(-3, 3)
plt.ylim(-10, 51)

plt.show()
````

### Заголовок (название графика)

````python
plt.plot(x, y1, 'b')
plt.plot(x, y2, 'r-.')
plt.plot(x, y3, 'g--')

plt.title('Графики matplotlib', fontsize = 15) # меняем размер
plt.show()
````

### Аннотации на графике

Команда ```plt.annotate('текст', xy, xytext, arrowprops)``` призвана создавать аннотации:  
* 'текст' - текст метки
* xy - местоположение метки
* xytext - местоположение указателя
* arrowprops - визуализация

````python
plt.plot(x, y1, 'b')
plt.plot(x, y2, 'r-.')
plt.plot(x, y3, 'g--')

plt.annotate('local max',
             xy = (4.8, 55),
             xytext = (-2, 75),
             arrowprops = {'facecolor': 'orange'},
             fontsize = 15)

plt.show()
````

### Сетка из графиков

В анализе данных нередко применяется создание **множества графиков** в виде сетки  
определенного размера, где каждый расположен в своей ячейке.  
Для построение такой сетки используется функция:  
```plt.subplot(кол-во строк, кол-во столбцов, номер ячейки)```  

Также возможно модернизированная версия с помощью GridSpec

````python
plt.figure(figsize = (15, 5))

x = np.arange(-5, 5, 0.1)

for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.plot(x, x**i)

plt.show()
````

### Гистрограмма

````python
x = np.random.randint(1, 5, 10)

plt.hist(x, color='r', orientation = 'horizontal') # нужная команда для создания гистограмм

plt.show()
````

### Столбчатая диаграмма

````python
temperature_2019 = np.random.randint(15, 25, 31)
temperature_2018 = np.random.randint(15, 25, 31)

x = np.arange(1, 32)

plt.figure(figsize = (15, 5))

plt.bar(x, temperature_2019, alpha = 0.7) # нужные команды для создания и дальнейшего наложения диаграмм
plt.bar(x, temperature_2018, alpha = 0.5)

plt.show()
````
