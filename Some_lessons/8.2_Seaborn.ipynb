### Seaborn - библиотека для более сложной визуализации

Seaborn — это библиотека для создания статистических графиков на Python. Она основывается на matplotlib и тесно взаимодействует со структурами данных pandas

````python
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
# sns.set_style('darkgrid') # изменение стиля
# sns.set_palette('tab10') # изменение палитры
````

### Диаграмма рассеяния — Scatter Plot

Диаграмма рассеяния — это диаграмма, которая отображает точки на основе двух измерений набора данных

````python
flights = sns.load_dataset('flights')
flights.head()
````

````python
sns.scatterplot(data = flights, x = 'year', y = 'passengers')

plt.show()
````

### Линейный график — Line Plot

Этот график рисует линию, которая представляет собой развитие непрерывных или категориальных данных

````python
sns.lineplot(data = flights, x = 'year', y = 'passengers')

plt.show()
````

### Столбчатая диаграмма — Bar Plot

Наверное, это самый известный тип диаграммы

````python
sns.barplot(data = flights, x = 'year', y = 'passengers')

plt.show()
````

### Расширение функционала с matplotlib

Seaborn основывается на matplotlib, расширяя ее функциональные возможности и абстрагируя сложность. При этом seaborn не теряет в своей мощности.  
Любая диаграмма seaborn может быть кастомизирована с помощью функций из библиотеки matplotlib. Допустим, мы хотим построить несколько диаграмм одновременно с помощью seaborn,  
в этом случае можно воспользоваться функцией ```subplot``` из matplotlib

````python
diamonds = sns.load_dataset('diamonds')

plt.subplot(1, 2, 1)
sns.countplot(x = 'carat', data = diamonds)

plt.subplot(1, 2, 2)
sns.countplot(x = 'depth', data = diamonds)

plt.show()
````

### Пример работы с датасетом

````python
tips_df = sns.load_dataset('tips')
tips_df.head()
````

````python
tips_df['tip_percentage'] = tips_df['tip'] / tips_df['total_bill']
tips_df.head()
````

### Процент чаевых

Большинство клиентов оставляют чаевые в размере от 15 до 20% от счета, но есть несколько случаев, когда чаевые превышают 70%.  
Эти значения называются аномалиями или выбросами, и на них всегда стоит обращать внимание, чтобы понять являются ли эти значения ошибочными

````python
sns.histplot(tips_df['tip_percentage'], binwidth = 0.05) 
# параметр 'пропускная способность' делает график более читабельным
plt.show()
````

На этот раз мы создали диаграмму с помощью всего набора данных, а не одного столбца, и установили свойство ```hue``` на столбец ```time```.  
Так в диаграмме будут использоваться разные цвета для каждого значения времени, и к ней добавится легенда

````python
sns.histplot(data = tips_df, x = 'tip_percentage', binwidth = 0.05, hue = 'time') 

plt.show()
````
