#### Задание 24

Выясните, у каких фильмов парный год номинации, например, 1927/28. Уберите вторую дату. Для примера получится 1927

```python
df['Oscar Year'] = df['Oscar Year'].astype('string').str.replace('/', '') # меняю тип 'object' на 'string' и убираю '/'
df['Oscar Year'] = df['Oscar Year'].astype('int') # меняю тип 'string' на 'int'

def numChange(x):
    if x > 190000:
        return int(x / 100)
    else:
        return x

df['Oscar Year'] = df['Oscar Year'].apply(numChange)
df
```

#### Задание 25

В датасете есть столбец года проведения церемонии Оскар. Добавьте к году дату - 25 апреля. Итоговый столбец должен иметь календарный тип данных.

```python
df['Oscar Year'] = df['Oscar Year'].astype('string') # преобразовали int в sting

df['Oscar Year'] = df['Oscar Year'].apply(lambda x: x + '0425') # добавили 04-25

df['Oscar Year'] = df['Oscar Year'].apply(lambda x: datetime.datetime.strptime(x, '%Y%m%d')) # преобразовал строку к datetime
```

#### Задание 26

Отсортируйте датасет по времени прохождения церемонии. С помощью оконных функций выведите средний рейтинг фильмов в каждом промежутке длиной 10 лет.

```python
df.sort_values(by = 'Oscar Year', ascending = False, inplace = True)

print('1927-1937: ', df[(df['Oscar Year'] > '1927') & (df['Oscar Year'] <= '1937')]['IMDB Rating'].mean())
print('1937-1947: ',df[(df['Oscar Year'] > '1937') & (df['Oscar Year'] <= '1947')]['IMDB Rating'].mean())
print('1947-1957: ',df[(df['Oscar Year'] > '1947') & (df['Oscar Year'] <= '1957')]['IMDB Rating'].mean())
print('1957-1967: ',df[(df['Oscar Year'] > '1957') & (df['Oscar Year'] <= '1967')]['IMDB Rating'].mean())
print('1967-1977: ',df[(df['Oscar Year'] > '1967') & (df['Oscar Year'] <= '1977')]['IMDB Rating'].mean())
print('1977-1987: ',df[(df['Oscar Year'] > '1977') & (df['Oscar Year'] <= '1987')]['IMDB Rating'].mean())
print('1987-1997: ',df[(df['Oscar Year'] > '1987') & (df['Oscar Year'] <= '1997')]['IMDB Rating'].mean())
print('1997-2007: ',df[(df['Oscar Year'] > '1997') & (df['Oscar Year'] <= '2007')]['IMDB Rating'].mean())
print('2007-2017: ',df[(df['Oscar Year'] > '2007') & (df['Oscar Year'] <= '2017')]['IMDB Rating'].mean())
print('2017+    : ',df[(df['Oscar Year'] > '2017')]['IMDB Rating'].mean())
```

#### Задание 27

Выведите три компании (столбец Production Company), фильмы которых завоевали самое большое количество наград

```python
dfAward = df.copy()

dfAward['Award'] = dfAward['Award'].astype('string')
dfAward['Award'] = dfAward['Award'].apply(lambda x: '1' if x == 'Winner' else '0')
dfAward['Award'] = dfAward['Award'].astype('int')

groupedPC = dfAward.groupby('Production Company')
groupedPC.sum(numeric_only = True).sort_values(by = 'Award', ascending = False).head(3)
```
